```python
from pulp import *
import pandas as pd
```

```python
# Define the model
model = LpProblem("Production Routing", LpMinimize)

# Define decision variables
x11 = LpVariable("x11", lowBound=0)
x12 = LpVariable("x12", lowBound=0)
x21 = LpVariable("x21", lowBound=0)
x22 = LpVariable("x22", lowBound=0)

# Define objective function
model += 3 * x11 + 5 * x12 + 2.5 * x21 + 4 * x22, "Total production cost"

# Define constraints
model += x11 + x12 >= 500, "Product 1 minimum requirement"
model += x21 + x22 >= 700, "Product 2 minimum requirement"
model += x11 + x21 <= 800, "Modern line capacity"
model += x12 + x22 <= 600, "Old line capacity"

# Solve the problem
status = model.solve()

# Print the results
print("Status:", LpStatus[status])
print("Minimum production cost:", round(model.objective.value(), 2))

for v in model.variables():
    print(v.name, "=", round(v.varValue, 2))
```
