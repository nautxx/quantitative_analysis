```python
from pulp import *
import pandas as pd
```

```python
# Create a LP minimization model
model = pulp.LpProblem("Diet", LpMinimize)

# Define decision variables
x = LpVariable('x', lowBound=0, upBound=2, cat='Continuous')  # pounds of Food A used per resident
y = LpVariable('y', lowBound=0, cat='Continuous')  # pounds of Food B used per resident

# Objective function to minimize cost
model += 35 * x + 7 * y, "Total Cost"

# Constraints
model += 600 * x + 900 * y >= 1800, "Calorie Lower Bound"
model += 600 * x + 900 * y <= 3600, "Calorie Upper Bound"
model += 200 * x + 700 * y <= 1400, "Starch Constraint"
model += 400 * x + 100 * y >= 400, "Protein Constraint"

# Solve the model
model.solve()

# Print the results
print("Optimal Solution:")
for var in model.variables():
    print(f"{var.name}: {var.varValue} pounds")
    
print(f"Total Cost: ${value(model.objective)}")
```
