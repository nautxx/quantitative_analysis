```python
from pulp import *
import pandas as pd
```

```python
# Define the problem
model = LpProblem("Budget Optimization", LpMaximize)

# Define decision variables
x1 = LpVariable("Food", lowBound=0)  # Amount spent on food
x2 = LpVariable("Shelter", lowBound=0)  # Amount spent on shelter
x3 = LpVariable("Entertainment", lowBound=0)  # Amount spent on entertainment

# Define objective function
model += 2 * x1 + 3 * x2 + 5 * x3

# Define constraints
model += x1 + x2 + x3 <= 30000  # Total budget constraint
model += x1 + x2 <= 22500  # Food and shelter combined constraint
model += x2 <= 4500  # Shelter constraint
model += x3 <= 13500  # Entertainment constraint

# Solve the problem
model.solve()

# Print the results
print("Optimal Solution:")
for v in model.variables():
    print(v.name, "=", v.varValue)

print("\nTotal satisfaction value:", value(model.objective))
```
